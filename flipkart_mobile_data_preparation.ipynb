```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Flipcart_Mobile.csv")
df.columns
```

```python
df.dtypes
```

```python
df.head()
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df = df.rename(columns={"Model Name": "Model_Name"})
```

```python
df['Price'] = df['Price'].fillna(df['Price'].median())
df['Avg Rating'] = df['Avg Rating'].fillna(df['Avg Rating'].median())
```

```python
df.isnull().sum()
```

```python
df['Rating Count'] = df['Rating Count'].str.replace(',', '').str.extract('(\d+)').astype(float)
df['Review Count'] = df['Review Count'].str.replace(',', '').str.extract('(\d+)').astype(float)
df[['Rating Count', 'Review Count']] = df[['Rating Count', 'Review Count']].fillna(0)
```

```python
df['Brand'] = df['Model_Name'].str.split(' ').str[0]
```

```python
df['Discount'] = df['Discount'].fillna('0% off')
```

```python
df['Discount'] = df['Discount'].str.replace('% off', '', regex=False).astype(float) / 100
```

```python
df['Battery_Capacity_mAh'] = df['Battery Capacity'].str.extract(r'(\d+)\s*mAh').astype(float)
```

```python
median_battery = df['Battery_Capacity_mAh'].median()
df['Battery_Capacity_mAh'] = df['Battery_Capacity_mAh'].fillna(median_battery)
```

```python
df
```

```python
df = df.drop(columns=['Battery Capacity'])
```

```python
df['Lens Detail'] = df['Lens Detail'].fillna('Not Specified')
df['Processor'] = df['Processor'].fillna('Unknown')
df['Warranty'] = df['Warranty'].fillna('Not Specified')
```

```python
df['Memory Details'] = df['Memory Details'].fillna('Not Specified')
```

```python
df = df.rename(columns={"Memory Details": "Memory_Details"})
```

```python
df = df.rename(columns={"Lens Detail": "Lens_Detail"})
```

```python
df = df.rename(columns={"Avg Rating": "Avg_Rating"})
```

```python
df = df.rename(columns={"Rating Count": "Rating_Count"})
```

```python
df = df.rename(columns={"Review Count": "Review_Count"})
```

```python
df = df.rename(columns={"Screen Dimention": "Screen_Dimention"})
```

```python
df.to_csv("Flipcart_Mobile_final_cleaned.csv", index=False)
```
